```python
import streamlit as st
import plotly.express as px
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
import_data = pd.read_csv('/home/umakanth/analytics/2021_MPG_data.csv')
```

```python
numerical_columns = []
categorical_columns = []

for i in import_data.columns:
    if import_data[i].dtype == np.dtype('float64') or import_data[i].dtype == np.dtype('int64'):
        numerical_columns.append(import_data[i])
    else:
        categorical_columns.append(import_data[i])
```

```python
numerical_data = pd.concat(numerical_columns, axis = 1)
categorical_data = pd.concat(categorical_columns, axis = 1)
```

```python
numerical_data = numerical_data.apply(lambda x:x.fillna(np.mean(x)))
```

```python
scaler = StandardScaler()
scaled_values = scaler.fit_transform(numerical_data)
```

```python
pca = PCA()
pca_data = pca.fit_transform(numerical_data)
pca_data = pd.DataFrame(pca_data)
pca_data
```

```python
new_column_names = ['PCA_' + str(i) for i in range(1,len(pca_data.columns)+1)]
new_column_names
```

```python
print(dict(zip(list(pca_data.columns),new_column_names)))
```

```python
column_mapper = dict(zip(list(pca_data.columns),new_column_names))
column_mapper
```

```python
pca_data = pca_data.rename(columns=column_mapper)
pca_data
```
